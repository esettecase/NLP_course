```python
len(emma_words)/len(emma_sentences) # calculate the average number of words per sentence in the text file austen-emma.txt
```

```python
from nltk.corpus import gutenberg
for fileid in gutenberg.fileids(): # for loop: iterate over the values of the file identifiers
    num_words = len(gutenberg.words(fileid))
    num_sentences = len(gutenberg.sents(fileid))
    print(num_words/num_sentences, fileid) # calculate the average number of words per sentence in alltext files of Gutenberg corpus
```

## Frequency distribution of a token

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(data)
```

```python
from nltk.probability import FreqDist
fdist=FreqDist(tokens)
fdist
```

```python
fdist['rich']
fdist['Emma']
```

```python
# Plot:
import matplotlib.pyplot as plt
fdist.plot(30,cumulative=False)
```
